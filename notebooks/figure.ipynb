```python
report20 = pd.read_csv("report_GNN_node_features20.txt",sep=',',skiprows=[0,1,2], header=None)
report50 = pd.read_csv("report_GNN_node_features50.txt",sep=',',skiprows=[0,1,2], header=None)
report100 = pd.read_csv("report_GNN_node_features100.txt",sep=',',skiprows=[0,1,2], header=None)
report20[1].str.split()
```

```python
plt.plot([k for k in range(len(report20[1].str.split()))],[float(k[2]) for k in report20[1].str.split()], label = "20 epoch")
plt.plot([k for k in range(len(report50[1].str.split()))],[float(k[2]) for k in report50[1].str.split()], label = "50 epoch")
plt.plot([k for k in range(len(report100[1].str.split()))],[float(k[2]) for k in report100[1].str.split()], label = "100 epoch")
plt.legend()
plt.title("L1 Loss for GNN ")
plt.show()
```
